```python
import requests

api_url = "https://apiest/products"

params = {
    "product_type": "smartphone"
}

response = requests.get(api_url, params=params)

if response.status_code == 200:
    data = response.json()
    
    for product in data:
        print(product)
else:
    print(f"การเรียก API ล้มเหลว (status code: {response.status_code})")
```

```python
import requests

api_url = "https://apiest/products"  

data = {"product_type": "smartphone"}  
 
response = requests.get(api_url, params=data)

if response.status_code == 200:
    print("การเรียก API สำเร็จ")
    api_data = response.json()

    if "product" in api_data:
        products = api_data["product"]
        
        producttype = None
        for product in products:
            if product['product_type'] == "smartphone":
                producttype = product
                break

        if producttype:
            print(f"Product : {producttype['product_type']}, Probability: {producttype['probability']}")
        else:
            print("ไม่พบ Product: smartphone ")
    else:
        print("ข้อมูลที่ส่งกลับไม่มีรายการ")
else:
    print(f"การเรียก API ล้มเหลว (status code: {response.status_code})")
```

```python
import requests

def orders_by_user_id():
    api_url = 'https://apitest/orders'  

    user_id = 12345

    params = {'user_id': user_id}
    response = requests.get(api_url, params=params)

   
    if response.status_code == 200:
        orders = response.json()
        if orders:
            print("สินค้าของ", user_id, ":")
            for order in orders:
                print(f"Order ID: {order['order_id']}, Product: {order['product_name']}, Quantity: {order['quantity']}")
        else:
            print("ไม่มีสินค้้าในตระกร้่า", user_id)
    else:
        print("ไม่สามารถค้นหาได้:", response.status_code)

if __name__ == '__main__':
    orders_by_user_id()
```

```python
import requests

def delete_product():
    api_url = 'https://apitest/delete' 
    user_id = 12345
    product_name = 'iPhone 13'  
    params = {'user_id': user_id, 'product_name': product_name}

    response = requests.delete(api_url, params=params)

    if response.status_code == 204:
        print(f"Product '{product_name}' ลบสินค้าสำเร็จ {user_id}")
    elif response.status_code == 404:
        print(f"Product '{product_name}' ไม่พบสสินค้าที่ต้องการลบ {user_id}")
    else:
        print(f"ลบไม่สำเร็จ '{product_name}' from the cart. Status code:", response.status_code)

if __name__ == '__main__':
    delete_product()
```
